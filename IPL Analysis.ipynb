```python
def runs_scored(batsman_name):
    runs_scored_by_batsman = delivery[delivery['batsman'] == batsman_name]
    return(runs_scored_by_batsman.groupby('bowling_team')['batsman_runs'].sum().sort_values(ascending = False).head(1))
```

```python
runs_scored('DA Warner')
```

```python
#find the most destructive death over(last 5 overs) batsman in the history of IPL.
#on basis of strike rate(SR = (number of runs/number of balls)/100 )
#min batsman 200 balls in over 16-20.
death_over = delivery[delivery['over'] > 15]
```

```python
all_batsman = death_over.groupby('batsman')['batsman_runs'].count() 
x = all_batsman > 200
all_batsman[x]
```

```python
batsman_list = all_batsman[x].index.tolist()
```

```python
final = delivery[delivery['batsman'].isin(batsman_list)]
```

```python
runs = final.groupby('batsman')['batsman_runs'].sum()
balls = final.groupby('batsman')['batsman_runs'].count()
```

```python
sr = (runs/balls)*100
sr.sort_values(ascending = False).head(1).index[0]
```
